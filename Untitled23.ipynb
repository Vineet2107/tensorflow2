```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
df = pd.read_csv('/content/drive/MyDrive/air_quality.csv')
```

```python
df.head()
```

```python
df['PM2.5']
```

```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['wind_direction'])
print(one_hot_encoded_data.head())
```

```python
one_hot_encoded_data.head()
```

```python
one_hot_encoded_data.drop(['year','month', 'day', 'hour'], axis=1, inplace=True)
```

```python
one_hot_encoded_data.head()
```

```python
df_pred=one_hot_encoded_data.loc[df['PM2.5'] == -1]
```

```python
df_pred.head()
```

```python
x_pred=df_pred.drop(['PM2.5'],axis=1)
```

```python
df_use=one_hot_encoded_data[one_hot_encoded_data['PM2.5'] != -1]
```

```python
df_use.head()
```

```python
training_data = df_use.sample(frac=0.8, random_state=25)
testing_data = df_use.drop(training_data.index)
```

```python
training_data.head()
```

```python
y_train=training_data['PM2.5']
y_train.head()
```

```python
x_train=training_data.drop(['PM2.5'], axis=1)
```

```python
x_train.head()
```

```python
y_test=testing_data['PM2.5']
y_test.head()
```

```python
x_test=testing_data.drop(['PM2.5'], axis=1)
x_test.head()
```

```python
model = keras.Sequential(
    [
     Dense(32, input_dim=20, activation='relu'),
     Dense(64, activation='relu'),
     Dense(128, activation='relu'),
     Dense(1, activation='relu')

    ]
)
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.001),
              loss=keras.losses.mae,
              )
```

```python
history = model.fit(
    x_train,
    y_train,
    batch_size=512,
    validation_data=(x_test, y_test),
    epochs=10,
)
```
